**Importing Libraries **

```python
# Importing pandas and MLXtednd
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

**Loading the Dataset**

```python
# Loading Dataset into the notebook with help of pandas and defining as df
df = pd.read_csv("Global_Superstore2.csv", encoding = 'unicode_escape', engine ='python')
df.head(5)
```

```python
df['Product Name'] = df['Product Name'].str.strip()
df.dropna(axis=0, subset=['Order ID'], inplace=True)
df['Order ID'] = df ['Order ID'].astype('str')
df = df[~df['Order ID'].str.contains('C')]
```

**Taking Region and Creating Basket**

```python
# Creating a Bakset with Region As State
basket = (df[df['State'] =="New York"]
          .groupby(['Order ID', 'Product Name'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('Order ID'))
```

```python
# Defining a Hot Encode
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
basket_sets
```

```python
# Iteration code for Iterrows
for i, j in df.iterrows():
  print(i, j)
  print()
```

```python
# Finding Support Value by Items with Apriori
frequent_itemsets = apriori(basket_sets, min_support=0.010, use_colnames=True)
frequent_itemsets.head(50)
```

```python
# Defingin Our Association Rule
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

```python
print(frequent_itemsets)
```

```python
# Finding Lift and Confidence Value
rules[ (rules['lift'] >= 2) &
       (rules['confidence'] >= 0.5) ]
```

**Results**

```python
# Specifing Association Rule to Find all the Market Analysis
association_rules(frequent_itemsets, metric="lift", min_threshold=1).sort_values('lift',ascending=False).reset_index(drop=True)
```
